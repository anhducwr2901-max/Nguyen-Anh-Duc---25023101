```python
a ,b = map(int, input().split())
if a > b:
    x = b
else:
    x = a
for i in range(1, x + 1):
    if a % i == 0 and b % i == 0:
        print(i)
```

Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
for i in range(1, n):
    if n == 2**i:
        print(f"{n} là số nguyên dương có dạng 2^k")
        break
```

Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
total = 0
while n > 0:
    total += n % 10
    n //= 10
print(total)
```

Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
countle = 0
countchan = 0
while n > 0:
    duoi = n % 10
    if duoi % 2 == 0:
        countchan += 1
    else:
        countle += 1
    n //= 10
print(f"Số lượng chữ số chẵn: {countchan}")
print(f"Số lượng chữ số lẻ: {countle}")
```

Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
total = 0
for i in range(1, n + 1):
    if total < n:
        total += i
    if total >= n:
        print(i - 1)
        break
```

Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
tong = 0
for i in range(1, 999**99):
    if tong <= A:
        tong += i**-1
    if tong > A:
        print(i)
        break
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
n = int(input())
if n == -1:
    pass
else:
    max = n
    min = n
    while True:
        n = int(input())
        if n == -1:
            break
        if n > max:
            max = n
        if n < min:
            min = n
    print(max)
    print(min)
```

[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 

```python
A = int(input())
f1, f2 = 1, 1
result = 1
while f2 <= A:
    result = f2
    f1, f2 = f2, f1 + f2
print(result)
```

Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input()
first_word = ""
for i in range(len(s)):
    if s[i] != ' ':
        first_word += s[i]
    elif first_word:
        break
print(first_word)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input()
inhoa = inthuong = so = 0
for ch in s:
    if ch.isupper():
        inhoa += 1
    elif ch.islower():
        inthuong += 1
    elif ch.isdigit():
        so += 1
print(f"Có {inhoa} ký tự in hoa")
print(f"Có {inthuong} ký tự in thường")
print(f"Có {so} ký tự số")
```
