Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
tu_dien = ["Không", "Một", "Hai", "Ba", "Bốn", "Năm", "Sáu", "Bảy", "Tám", "Chín"]

so = int(input("Nhập số từ 0 đến 9: "))

if 0 <= so <= 9:
    print(tu_dien[so])
else:
    print("Nhập sai")
```

```python
so = int(input("Nhập một số từ 0 đến 9: "))

match so:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    
    case _: 
        print("Nhập sai")
```
