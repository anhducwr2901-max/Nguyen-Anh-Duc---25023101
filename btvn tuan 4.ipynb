```python
#Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

n = int(input("Nhập số nguyên dương n: "))

steps = 0

while True:
    chuoi_n = str(n)
    
    chuoi_dao_nguoc = chuoi_n[::-1]
    
    if chuoi_n == chuoi_dao_nguoc:
        break  
    
    so_dao_nguoc = int(chuoi_dao_nguoc)
    
    n = n + so_dao_nguoc
    
    steps = steps + 1

print("Số bước:", steps)
print("Giá trị palindrome thu được:", n)
```

Nhập vào số nguyên dương n. In ra tất cả các số chính phương <= n có các chữ số đôi một khác nhau( in theo thứ tự tăng dần, cùng hàng, cách nhau một dấu cách). Nếu không có   , in ra no number

```python
n = int(input("Nhập n: "))
ans=[]
for i in range(1,int(n**0.5)+1):
    square = i*i
    s = str(square)

    if len(set(s)) == len(s):
        ans.append(square)

if not ans:
    print("no number")
else:
    print(*ans)
```

"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 

```python
n = int(input())

MaxL = 0
Minx = 1

for i in range(1, n + 1):
    x = i
    current_L = 1 
    
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3 * x + 1
        current_L += 1
    
    if current_L > MaxL:
        MaxL = current_L
        Minx = i

print(Minx, MaxL)
```

[ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a, b = map(int, input().split())

def sum_divisors(n):
    if n <= 1:
        return 0
    total = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            # Cộng thêm ước chéo nếu khác nhau (VD: n=20, i=4 thì cộng cả 4 và 5)
            if i != n // i:
                total += n // i
    return total

if a != b and sum_divisors(a) == b and sum_divisors(b) == a:
    print("true")
else:
    print("false")
```

[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
m , n = map(int, input().split())
uoc_chung = []
for a in range(1, min(m, n) + 1):
    if m % a == 0 and n % a == 0:
        uoc_chung.append(a)
c = max(uoc_chung)
print(c)
```
