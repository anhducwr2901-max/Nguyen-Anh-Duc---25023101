Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.

```python
a = int(input())
pi = 3.14
s = pi * a * a
chuvi = 2 * pi * a
print("Dien tich hinh tron la:", s)
print("Chu vi hinh tron la:", chuvi)
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c = map(int, input().split())
if a + b > c and a + c > b and b + c > a:
    print("Day la 3 canh cua 1 tam giac")
else:
    print("Day khong phai la 3 canh cua 1 tam giac")
if a == b == c:
    print("Day la 3 canh cua 1 tam giac deu")
elif a == b or a == c or b == c:
    print("Day la 3 canh cua 1 tam giac can")
else:
    print("Day la 3 canh cua 1 tam giac thuong")
```

Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0) 

```python
a = int(input())
b = int(input())
if a == 0 and b == 0:
    print("Phuong trinh co vo so nghiem")
elif a == 0 and b != 0:
    print("Phuong trinh vo nghiem")
else:
    x = -b / a
    print("Phuong trinh co nghiem duy nhat x =", x)
```

Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

```python
a,b,c,d = map(int, input().split())
max_value = max(a, b, c, d)
print("Gia tri lon nhat la:", max_value)
```

Bài 7: Tìm giá trị nhỏ nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

```python
x = int(input("Nhập vào số giây: "))

gio = x // 3600
phut = (x % 3600) // 60
giay = x % 60

print(f"{x} giây bằng: {gio} giờ, {phut} phút, {giay} giây")
```
